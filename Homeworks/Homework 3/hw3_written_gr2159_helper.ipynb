# Q1

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = logistic(x)

plt.plot(x, y)
plt.show()
```

```python
w0 = 2
y2 = logistic(x + w0)

plt.plot(x, y, label="w0=0")
plt.plot(x, y2, label=f"w0={w0}")
plt.legend()
plt.show()
```

```python
w0 = -2
y3 = logistic(x + w0)

plt.plot(x, y, label="w0=0")
plt.plot(x, y3, label=f"w0={w0}")
plt.legend()
plt.show()
```

```python
w_prime = 2 * x
y3 = logistic(w_prime)

plt.plot(x, y, label="w")
plt.plot(x, y3, label=f"w'")
plt.legend()
plt.show()
```
